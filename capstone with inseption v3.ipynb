# importing requried library

```python
from tensorflow.keras.layers import Input, Dense, Flatten
from tensorflow.keras.models import Model
from tensorflow.keras.applications.inception_v3 import InceptionV3
#from keras.applications.vgg16 import VGG16
from tensorflow.keras.applications.inception_v3 import preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img
from tensorflow.keras.models import Sequential
import numpy as np
from glob import glob
```

# assigning image size and train and validation data path

```python
IMAGE_SIZE = [224, 224]

train_path = r'C:\Users\KIRUBA KARAN\Desktop\python\01.capstone project2\resize image\train'
valid_path = r'C:\Users\KIRUBA KARAN\Desktop\python\01.capstone project2\resize image\valid'
```

# assigning input object for inspection v3 model

```python
inception = InceptionV3(input_shape=IMAGE_SIZE + [3], weights='imagenet', include_top=False)
```

```python
folders = glob(r'C:\Users\KIRUBA KARAN\Desktop\python\01.capstone project2\resize image\train/*')
```

# for using all the weights

```python
for layer in inception.layers:
    layer.trainable = False
```

# assigning flatten layer

```python
x = Flatten()(inception.output)
```

# assigning the output layer and building the inspection v3 model

```python
prediction = Dense(len(folders), activation='sigmoid')(x)

# create a model object
model = Model(inputs=inception.input, outputs=prediction)
```

```python
model.summary()
```

# compiling the model with loss,optimiser and metrics

```python
model.compile(
  loss='binary_crossentropy',
  optimizer='adam',
  metrics=['accuracy']
)
```

# importing image geneartor and images are imported through flow from directory for both test and train image

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)
```

```python
valid_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
training_set = train_datagen.flow_from_directory(train_path,
                                                 target_size = (224, 224),
                                                 batch_size = 32,
                                                 class_mode = 'binary')
```

```python
valid_set = valid_datagen.flow_from_directory(valid_path,
                                            target_size = (224, 224),
                                            batch_size = 32,
                                            class_mode = 'binary')
```

# fitting the model with train and valid data 

```python
r = model.fit(
  training_set,
  validation_data=valid_set,
  epochs=10,
  steps_per_epoch=len(training_set),
  validation_steps=len(valid_set)
)
```

# ploting the accuracy and loss of both train and test

```python
import matplotlib.pyplot as plt
plt.plot(r.history['loss'], label='train loss')
plt.plot(r.history['val_loss'], label='val loss')
plt.legend()
plt.show()
plt.savefig('LossVal_loss')

# plot the accuracy
plt.plot(r.history['accuracy'], label='train acc')
plt.plot(r.history['val_accuracy'], label='val acc')
plt.legend()
plt.show()
plt.savefig('AccVal_acc')
```
